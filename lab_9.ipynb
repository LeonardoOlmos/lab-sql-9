```python
import pandas as pd
from sqlalchemy import create_engine, text
import pymysql
import getpass
```

```python
# Define connection URL

url = f'mysql+pymysql://root:{password}@localhost/sakila'

engine = create_engine(url)
```

```python
""" 7. Create a Python connection with SQL database and retrieve the results of the last two queries (also mentioned below) as dataframes:
	- Check the number of rentals for each customer for May
	- Check the number of rentals for each customer for June
	Hint: You can store the results from the two queries in two separate dataframes.
"""

df_rentals_may = pd.read_sql_query('SELECT customer_id, COUNT(*) AS total_rentals \
FROM rentals_may \
GROUP BY customer_id;', engine);

df_rentals_june = pd.read_sql_query('SELECT customer_id, COUNT(*) AS total_rentals \
FROM rentals_june \
GROUP BY customer_id;', engine);
```

```python
""" 8. Write a function that checks if customer borrowed more or less films in the month of June as compared to May.
  Hint: For this part, you can create a join between the two dataframes created before, using the merge function available for pandas dataframes. 
"""

# Merging rentals dataframes
df_rentals = df_rentals_may.merge(df_rentals_june, on = ['customer_id'], suffixes = ('_may', '_june'))
```

```python
df_rentals.head(2)
```

```python
# Function to compare rentals on june vs may
def customer_borrows(customer_id, rentals_may, rentals_june):
    if rentals_may > rentals_june:
        return f'Customer: {customer_id}, has borrowed more movies on May({rentals_may}) than June({rentals_june})'
    elif rentals_may < rentals_june:
        return f'Customer: {customer_id}, has borrowed more movies on June({rentals_june}) than May({rentals_may})'
    else:
        return f'Customer: {customer_id}, has borrowed the same movies on June and May'
```

```python
# loop through the rows using iterrows() and calling function to compare
for _, row in df_rentals.iterrows():
    print(customer_borrows(row['customer_id'], row['total_rentals_may'], row['total_rentals_june']))
```
